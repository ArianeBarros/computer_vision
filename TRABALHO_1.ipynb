```python
url = "https://a.tulv.in/uploads/screenshot-from-2021-02-23-19-04-26.png"
img_alpha = imread(url)

plt.title('Imagem Original')
plt.imshow(img_alpha)
plt.axis('off')
plt.show()
```

```python
img_alpha.shape #Pelo valor da terceira dimensão, sabemos que a imagem tem um canal alfa
```

```python
img = cv2.cvtColor(img_alpha, cv2.COLOR_BGR2GRAY).astype(np.uint8);
cv2_imshow(img)
```

```python
filtragem  = cv2.medianBlur(src=img,ksize=13)
cv2_imshow(filtragem)
```

```python
# Aplicando a transformação de intensidade Thresholding.
threshold_value = 0.4  # Valor do limiar

# image >= threshold_value: Essa é a condição que define o limiar de intensidade.
# Ela compara cada valor de intensidade na imagem image com o valor de limiarização threshold_value
# retorna um array de booleanos indicando se a condição é verdadeira ou falsa para cada pixel.

# Os pixels com intensidade maior ou igual ao limiar são considerados verdadeiros (True),
# enquanto os pixels com intensidade abaixo do limiar são considerados falsos (False).

# 1.0: É o valor atribuído aos pixels que atendem à condição (verdadeiros).
# 0.0: É o valor atribuído aos pixels que não atendem à condição (falsos).

thresholded_image_point_four = np.where(filtragem >= threshold_value, 1.0, 0.0)
```

```python
# Aplicando a transformação de intensidade Thresholding.
threshold_value = 0.5

thresholded_image_point_five = np.where(filtragem >= threshold_value, 1.0, 0.0)
```

```python
plt.title('Imagem Original')
plt.imshow(img_alpha)
plt.axis('off')
plt.show()

plt.title('Imagem filtrada com filtro Blur')
plt.imshow(filtragem)
plt.axis('off')
plt.show()
```

```python
# Plotando a imagem binarizada com limiar 0.4
plt.title('Plotando a imagem binarizada com limiar 0.4')
plt.imshow(thresholded_image_point_four)
plt.axis('off')
plt.show()

# Plotando a imagem binarizada com limiar 0.5
plt.title('Plotando a imagem binarizada com limiar 0.5')
plt.imshow(thresholded_image_point_five)
plt.axis('off')
plt.show()
```

```python
# Importações
import tensorflow as tf

# Carregando a base de dados
mnist = tf.keras.datasets.mnist

# Carregando o conjunto de treinamento e o conjunto de teste
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

## Execução 1

```python
# Estruturando a arquitetura da rede/ modelo
model = tf.keras.models.Sequential([
 tf.keras.layers.Flatten(),
 tf.keras.layers.Dense(784, activation=tf.nn.tanh),
 tf.keras.layers.Dense(393, activation=tf.nn.tanh),
 tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Compilando o modelo
## Utiliza o otimizador ADAM
## Função de perda: sparse_categorical_crossentropy
## Métrica empregada: accuracy

model.compile(optimizer='adam',
 loss='sparse_categorical_crossentropy',
 metrics=['accuracy'])

# Treinamento do modelo com 7 épocas
model.fit(x_train, y_train, epochs=7, verbose = 1)
```

```python
# Avaliação do modelo
model.evaluate(x_test, y_test)
```
